```python
# dropping two additional columns
df_qpp.drop (['state_fips', 'county_fips'], axis = 1, inplace = True)
```

```python
#rename columns with long names
df_qpp.rename(columns = {'number_of_fee_for_service_beneficiaries' : 'number_ffs_members',
                        'average_number_of_users_per_provider' : 'avg_user_per_provider',
                        'percentage_of_users_out_of_ffs_beneficiaries' : 'percent_users_out_of_ffs_members',
                        'average_number_of_providers_per_county' : 'avg_provider_per_county',
                        'percentage_of_dual_eligible_users_out_of_total_users' : 'percent_dual_elig_user_from_total',
                        'percentage_of_dual_eligible_users_out_of_dual_eligible_ffs_beneficiaries' : 'percent_dual_elig_user_from_dual_elig_ffs_member',
                        'number_of_fee_for_service_beneficiaries_change' : 'number_ffs_member_change',
                        'average_number_of_users_per_provider_change' : 'avg_number_user_per_provider_change',
                        'percentage_of_users_out_of_ffs_beneficiaries_change' : 'percent_user_out_ffs_member_change',
                        'average_number_of_providers_per_county_change' : 'avg_number_provider_per_county_change',
                        'percentage_of_dual_eligible_users_out_of_total_users_change' : 'percent_dual_elig_user_from_total_user_change',
                        'percentage_of_dual_eligible_users_out_of_dual_eligible_ffs_beneficiaries_change': "percent_dual_elig_user_from_dual_elig_ffs_member_change"}, inplace = True)
```

```python
#rename type of service to service_type
df_qpp.rename(columns = {'type_of_service' : 'service_type'}, inplace = True)
```

```python
#rename reference_period to year
df_qpp.rename(columns = {'reference_period' : 'year'}, inplace = True)
```

## Check data types, value formatting, and make corrections

```python
# check data types
df_qpp.dtypes
```

```python
# list of columns to remove commas
columns_to_clean = ['number_ffs_members', 'number_of_providers', 'number_of_users']

# remove commas from the selected columns
df_qpp[columns_to_clean] = df_qpp[columns_to_clean].applymap(lambda x: int(x.replace(',', '')))

# confirm the updated data types
df_qpp.dtypes
```

```python
# list of columns to remove commas to convert to float
columns_convert = ['avg_user_per_provider', 'avg_provider_per_county']

# remove commas and convert to float for the selected columns
df_qpp[columns_convert] = df_qpp[columns_convert].applymap(lambda x: float(x.replace(',', '')) if x.strip() != '' else 'NaN')

# confirm the updated data types
df_qpp.dtypes
```

```python
#change reference_period values to display only years
df_qpp['year'] = df_qpp['year'].str.split(' to ')

start_dates = pd.to_datetime(df_qpp['year'].str[0])
end_dates = pd.to_datetime(df_qpp['year'].str[1])

df_qpp['year'] = start_dates.dt.strftime('%Y')
```

```python
#confirm range was removed and only year remains
df_qpp['year'].head()
```

```python
#change dtype for year to int
df_qpp['year'] = df_qpp['year'].astype('int')
```

```python
#confirm update of dtype change
df_qpp.dtypes
```

```python
# remove dollar sign and commas from total payment and change dtype to float
df_qpp['total_payment'] = df_qpp['total_payment'].str.replace(',', '').str.replace('$', '').astype(float)
```

```python
#confirm removal of dollar sign and commas
df_qpp['total_payment'].head(15)
```
